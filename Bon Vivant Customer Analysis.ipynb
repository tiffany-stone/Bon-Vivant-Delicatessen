<h3>Overview</h3>
<p>Bon Vivant Delicatessen is a globally renowned retailer in the food industry.
<p>They sell products from 5 major categories with certain items considered Gold or exclusive products:</p>

<ul>
    <li>Wines</li>
    <li>Fruits</li>
    <li>Meat Products</li>
    <li>Fish Dishes</li>
    <li>Sweets</li>
</ul>

<p>Customers can order and acquire products through 3 sales channels:</p>

<ul>
    <li>Physical Stores</li>
    <li>Catalog</li>
    <li>Company Website</li>
</ul>

<a id="objective"></a>
<h3>Objective</h3>

<p>Bon Vivant wants to improve the performance of their marketing activities, with a special focus on marketing campaigns. The objective is to develop a BVD customer profile to target only those customers who would be receptive to future offerings while leaving out the non-respondents.
<br>
<p> An initial 5-part promotional campaign involving 2,240 customers was carried out.<br>
Customers were selected at random and contacted by phone.<br>During those months, socio-economic and firmographic data were collected to develop the Bon Vivant Delicatessen customer profile.</p></h5>

<h3>Questions this analysis aims to answer:</h3>

<h4>Products</h4>
<ul>
    <li>What are the high-demand products?</li>
    <li>What are the low-demand products?</li>
    <li>Are gold products a high-demand product?</li>
    <li>What type of customer purchases mostly gold products?</li>
</ul>

<h4>Place</h4>
<ul>
    <li>How are most purchases being made: in-store, catalog, website?</li>
    <li>Do young, single customers purchase from the website more than older customers?</li>
</ul>

<h4>Promotion</h4>
<ul>
    <li>Are discount purchases being utilized?</li>
    <li>Where are discount purchases being made (in-store, catalog, website)?</li>
    <li>Is there a product that is purchased most often through discount?</li>
    <li>Do older, single customers tend to take advantage of discount purchases?</li>
</ul>

<h4>Campaign Tests</h4>
<ul>
    <li>What was the conversion rate of each campaign?</li>
    <li>What is the customer profile for an early adopter?</li>
</ul>

<p>Z_CostContact and Z_Revenue have only one value, and ID is unnecessary in this analysis. These will not help gain any information on our customers so those columns will be removed.</p>

```python
# Remove unused columns
df.drop(columns=['ID', 'Z_CostContact', 'Z_Revenue'], inplace=True)
```

<h3>Can't do arithmetic operations with NaN, the result will always be NaN</h3>

```python
# Return the new dataframe length

# Return the row and column indices where the value is NaN:
# Remove NaN values

# Make a new dataframe with dropped values
new_data = df.dropna(axis=0, how='any')

print('Old dataframe length:', len(df),
      '\nNew dataframe length:', len(new_data),
      '\nNumber of rows that had NaN value:', (len(new_data)-len(df)))
```

```python
# Compare old dataframe with new dataframe

# Make a copy of old data frame
customers_df = new_data.copy()

# Create a value with all null values in new data frame
customers_df["Null Column"] = None

# Check if column is inserted properly
print(new_data.columns.values, "\n", customers_df.columns.values)

# Compare values before dropping null column
print("\nColumn number before dropping Null column\n",
	len(new_data.dtypes), len(customers_df.dtypes))

# Drop column with all null values
customers_df.dropna(axis=1, how='all', inplace=True)

# Compare values after dropping null column
print("\nColumn number after dropping Null column\n",
	len(new_data.dtypes), len(customers_df.dtypes))
```

<h4>Indentify Extreme Outliers</h4>

```python
# Identify extreme outliers in data (on numeric data)
customers_df[['income',
              'wine_amt',
              'fruit_amt',
              'meat_amt',
              'fish_amt',
              'sweet_amt',
              'gold_products',
              'num_discounts',
              'web',
              'catalog',
              'store',
              'mthly_web_visits',
              'total_spent']].describe()
```

<h3>Insights:</h3>

<p> The income variable has some extreme outliers that will be removed because they can signficantly skew the analysis. Wine, Meat, Fish, Gold Products, Sweet, and Fruit variables have outliers as well; however, they will not be discarded. There is no indication that they are data entry errors. They will be considered in the statistical analysis as they can portentially provide valuable information about the data distribution and possible insights into our customers.</p>

```python
# Visualize Income before outlier removal
sns.color_palette('rocket')
df_to_plot = customers_df['income']


# subplots
df_to_plot.plot(kind='box', color='purple')
plt.suptitle("Before Outlier Removal")
plt.show()
```

```python
# Visualize Income after outlier removal

# To remove outliers from multiple columns with different data types, will need to apply outlier removel separately for each data type.
cols = ['income']

numeric_cols_float = customers_df[cols].select_dtypes(include=np.float64).columns

# Calculate Q1, Q3, and IQR for float columns
Q1_float = customers_df[numeric_cols_float].quantile(.25, axis=0)
Q3_float = customers_df[numeric_cols_float].quantile(.75, axis=0)
IQR_float = Q3_float - Q1_float

# Calculate lower and upper limits for float columns
lower_limit_float = Q1_float - (1.5 * IQR_float)
upper_limit_float = Q3_float + (1.5 * IQR_float)

# Remove outliers from float columns
customers_df = customers_df[~((customers_df[numeric_cols_float] > upper_limit_float) | (customers_df[numeric_cols_float] < lower_limit_float)).any(axis=1)]


customers_df[cols].plot(kind='box', color='purple')
plt.suptitle("Boxplots After Outlier Removal IQR")
plt.show()
```

```python
# Customer Profile Results

import pandas as pd

# Columns to keep
columns_to_keep = ['dob', 'education', 'marital_status', 'child_home', 'teen_home', 'age', 'store', 'catalog', 'web']

# Select only the columns you want to keep
selected_columns = customers_df[columns_to_keep]

# Separate numeric and categorical columns
numeric_columns = selected_columns.select_dtypes(include=['int64', 'float64'])
categorical_columns = selected_columns.select_dtypes(include=['category'])

# Calculate mean for numeric columns
numeric_mean = pd.DataFrame(round(numeric_columns.mean(), 1), columns=['Average'])

# Calculate mode for categorical columns
categorical_mode = categorical_columns.mode().transpose()[0].to_frame(name='Mode')

# Combine results for both numeric and categorical columns
demographics = pd.concat([numeric_mean, categorical_mode])

# Reorder the rows in the desired order
demographics = demographics.reindex(columns_to_keep)


print(demographics)
```
